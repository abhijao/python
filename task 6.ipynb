```python
type_counts =df.type.value_counts()
```

```python
Questions = ["1. How many movies and TV shows are listed in the dataset?",
             "2. Which country has produced the highest number of titles?",
             "3. What is the distribution of titles based on their ratings?",
             "4. What are the top 5 directors who have directed the most titles?",
             "5. How has the number of titles released per year changed over time?"]
```

```python
# Ensure 'type_counts' is available before plotting
if 'type' in df.columns:
    plt.figure(figsize=(12, 4))
    sns.barplot(x=type_counts.index, y=type_counts.values, palette='viridis')
    plt.title('Number of Movies and TV Shows')
    plt.xlabel('Type')
    plt.ylabel('Count')
    plt.show()
```

```python
# Question 1: How many movies and TV shows are listed in the dataset?


plt.figure(figsize=(6, 6))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])
plt.title('Distribution of Movies and TV Shows')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Plot the top 10 countries by number of titles
country_counts = df['country'].value_counts().head(10)
plt.figure(figsize=(12, 4))
sns.barplot(x=country_counts.index, y=country_counts.values, palette='viridis')
plt.title('Top 10 Countries by Number of Titles')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Question 2: Which country has produced the highest number of titles?

country_counts = df['country'].value_counts().head(10)
plt.figure(figsize=(8, 8))
plt.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', len(country_counts)))
plt.title('Top 10 Countries by Number of Titles')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
#Plot the distribution of titles based on their ratings. 

plt.figure(figsize=(12, 4))
sns.countplot(data=df, x='rating', palette='viridis')
plt.title('Distribution of Titles by Rating')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Question 3: What is the distribution of titles based on their ratings?

rating_counts = df['rating'].value_counts()
plt.figure(figsize=(7, 7))
plt.pie(rating_counts, labels=rating_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', len(rating_counts)))
plt.title('Distribution of Titles by Rating')
plt.axis('equal')  
plt.show()
```

```python
# Plot the distribution of titles by rating

plt.figure(figsize=(12, 4))
sns.barplot(x=rating_counts.index, y=rating_counts.values, palette='viridis')
plt.title('Distribution of Titles by Rating')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Question 4: What are the top 5 directors who have directed the most titles?

directors = df['director'].dropna().str.split(',').explode()
top_directors = directors.value_counts().head(5)
plt.figure(figsize=(5, 5))
plt.pie(top_directors, labels=top_directors.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', len(top_directors)))
plt.title('Top 5 Directors by Number of Titles')
plt.axis('equal')  
plt.show()
```

```python
# Plot the top 5 directors

plt.figure(figsize=(12, 4))
sns.barplot(x=top_directors.values, y=top_directors.index, palette='viridis')
plt.title('Top 5 Directors by Number of Titles')
plt.xlabel('Count')
plt.ylabel('Director')
plt.show()
```
