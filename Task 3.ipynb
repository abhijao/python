<h2>Bar Plot

```python
import matplotlib.pyplot as plt
x=["python","c","c++","Java"]
y=[85,70,60,82]
plt.xlabel("Language")
plt.ylabel("No.")
plt.title("Bar Chart")

plt.bar(x,y,label="bar chart")
plt.legend()
plt.show()
```
